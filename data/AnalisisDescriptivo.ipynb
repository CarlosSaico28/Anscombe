```python
import pandas as pd
# Ruta al archivo CSV
ruta_csv = "C:/Users/carlo/Desktop/Mineria\Python\P1-Tableau\docentes-anonim-2018-filtrado.csv"
# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(ruta_csv, delimiter=',', encoding='latin1', low_memory=False)
# Imprimir el número de filas y columnas del DataFrame
print("El DataFrame tiene", df.shape[0], "filas y", df.shape[1], "columnas.")
print("El DataFrame filtrado tiene", df.shape[0], "filas y", df.shape[1], "columnas.")

# Visualizar las primeras filas del DataFrame
df.head(5)
```

```python
# Seleccionar las columnas específicas utilizando loc
df_subset = df.loc[:, ['codigo_ies', 'fecha_inicio_periodo', 'fecha_fin_periodo', 'sexo', 'nivel', 'nombre_ies_internacional', 'pais_estudio', 'nombre_titulo', 'cursando']]

# Imprimir el número de filas y columnas del DataFrame resultante
print("El DataFrame resultante tiene", df_subset.shape[0], "filas y", df_subset.shape[1], "columnas.")

# Visualizar las primeras filas del DataFrame resultante
df_subset.head(5)
```

##### df_subset.info()

```python
# Imprimir los niveles únicos
print("Niveles únicos en la columna 'nivel':")
print(df_subset['nivel'].unique())
```

```python
# Reemplazar los valores en la columna 'sexo'
df_subset['sexo'] = df_subset['sexo'].replace({'HOMBRE': 0, 'MUJER': 1})

# Reemplazar los valores en la columna 'nivel'
df_subset['nivel'] = df_subset['nivel'].replace({'CUARTO_NIVEL': 4, 'CUARTO_NIVEL_': 4, 'TERCER_NIVEL': 3})

# Reemplazar los valores en la columna 'cursando'
df_subset['cursando'] = df_subset['cursando'].replace({'NO': 0, 'SI': 1})

# Visualizar las primeras filas del DataFrame con las transformaciones
df_subset.head(5)
```

```python
df_subset.info()
```

```python
# Guardar el DataFrame modificado en un nuevo archivo CSV
ruta_nuevo_csv = "C:/Users/carlo/Desktop/Mineria\Python\P1-Tableau\docentes_normalizado.csv"
df_subset.to_csv(ruta_nuevo_csv, index=False)
print("El nuevo dataset ha sido guardado en:", ruta_nuevo_csv)
```

```python
import pandas as pd

# Ruta al archivo CSV
ruta_csv = "C:/Users/carlo/Desktop/Mineria\Python\P1-Tableau\docentes_normalizado.csv"

# Cargar el archivo CSV en un DataFrame
dfn = pd.read_csv(ruta_csv, delimiter=',', encoding='latin1', low_memory=False)

# Imprimir el número de filas y columnas del DataFrame
print("El DataFrame tiene", dfn.shape[0], "filas y", dfn.shape[1], "columnas.")

# Visualizar las primeras filas del DataFrame
dfn.head(5)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
# Contar los valores únicos en la columna 'cursando'
conteo = dfn['cursando'].value_counts()

# Crear la gráfica de barras
plt.bar(['0', '1'], conteo.values)

print("0-NO")
print("1-SI")

# Añadir etiquetas y título
plt.xlabel('Valor')
plt.ylabel('Cantidad')

# Añadir el número de 0s y 1s a la gráfica
for i, valor in enumerate(conteo.values):
    plt.text(i, valor, str(valor), ha='center', va='bottom')

plt.title('Cantidad de 0 y 1 en la columna "cursando"')

# Mostrar la gráfica
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Contar los valores únicos en la columna 'cursando'
conteo = dfn['sexo'].value_counts()

# Crear la gráfica de barras
plt.bar(['0', '1'], conteo.values)

# Añadir etiquetas y título
plt.xlabel('Valor')
plt.ylabel('Cantidad')

# Añadir el número de 0s y 1s a la gráfica
for i, valor in enumerate(conteo.values):
    plt.text(i, valor, str(valor), ha='center', va='bottom')

plt.title('Cantidad de hombres 0 y mujeres 1')

# Mostrar la gráfica
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Contar los valores únicos en la columna 'cursando'
conteo = dfn['nivel'].value_counts()

# Crear la gráfica de barras
plt.bar(['4', '3'], conteo.values)

# Añadir etiquetas y título
plt.xlabel('Valor')
plt.ylabel('Cantidad')

# Añadir el número de 0s y 1s a la gráfica
for i, valor in enumerate(conteo.values):
    plt.text(i, valor, str(valor), ha='center', va='bottom')

plt.title('Cantidad de 3 y 4 nivel')

# Mostrar la gráfica
plt.show()
```
